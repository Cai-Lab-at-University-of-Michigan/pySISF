```python
import pySISF

print(pySISF.__version__)

import numpy as np
```

```python
data = np.arange(2000**3, dtype='uint16').reshape(2000, 2000, 2000)
```

```python
pySISF.sisf.create_sisf(
    'test.sisf',
    data,
    data.shape,
    [256]*3,
    (1,1,1),
    compression=3,
    enable_status=True,
    thread_count=10,
)
```

```python
a = pySISF.sisf.sisf('test.sisf')
```
